```python
import pygame

# Initialize Pygame
pygame.init()

# Set the width and height of the screen [width, height]
size = (600, 600)
screen = pygame.display.set_mode(size)

pygame.display.set_caption("WarVessels")

# Colors
white = (255, 255, 255)
blue = (50, 153, 213)

# Font
font = pygame.font.Font(None, 36)

# Grid
grid = []
for row in range(10):
    grid.append([])
    for col in range(10):
        grid[row].append("O")

# Place ships on the grid for player 1
grid[0][0] = "X"
grid[1][1] = "X"
grid[2][2] = "X"

# Place ships on the grid for player 2
grid[5][5] = "X"
grid[6][6] = "X"
grid[7][7] = "X"

# Track whose turn it is
turn = 1

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEBUTTONUP:
            # Get the x, y coordinates of the mouse click
            x, y = pygame.mouse.get_pos()

            # Determine which grid cell was clicked
            row = y // 60
            col = x // 60

            # Check whose turn it is
            if turn == 1:
                # Check if the guess is correct for player 1
                if grid[row][col] == "X":
                    text = font.render("Hit!", True, (0, 255, 0))
                    screen.blit(text, (250, 300))
                    pygame.display.update()
                    pygame.time.wait(3000)
                    turn = 2
                else:
                    text = font.render("Miss!", True, (255, 0, 0))
                    screen.blit(text, (250, 300))
                    pygame.display.update()
                    pygame.time.wait(3000)
                    turn = 2
            elif turn == 2:
                # Check if the guess is correct for player 2
                if grid[row][col] == "X":
                    text = font.render("Hit!", True, (0, 255, 0))
                    screen.blit(text, (250, 300))
                    pygame.display.update()
                    pygame.time.wait(3000)
                    turn = 1
                else:
                    text = font.render("Miss!", True, (255, 0, 0))
                    screen.blit(text, (250, 300))
                    pygame.display.update()
                    pygame.time.wait(3000)
                    turn = 1

    # Clear the screen
    screen.fill((white))

    # Draw the grid
    for row in range(10):
        for col in range(10):
            pygame.draw.rect(screen, blue, (col * 60, row * 60, 60, 60),2)
    pygame.draw.line(screen, blue, (60, 0), (60, 600), 2)
    pygame.draw.line(screen, blue, (120, 0), (120, 600), 2)
    pygame.draw.line(screen, blue, (180, 0), (180, 600), 2)
    pygame.draw.line(screen, blue, (240, 0), (240, 600), 2)
    pygame.draw.line(screen, blue, (300, 0), (300, 600), 2)
    pygame.draw.line(screen, blue, (360, 0), (360, 600), 2)
    pygame.draw.line(screen, blue, (420, 0), (420, 600), 2)
    pygame.draw.line(screen, blue, (480, 0), (480, 600), 2)
    pygame.draw.line(screen, blue, (540, 0), (540, 600), 2)

    pygame.draw.line(screen, blue, (0, 60), (600, 60), 2)
    pygame.draw.line(screen, blue, (0, 120), (600, 120), 2)
    pygame.draw.line(screen, blue, (0, 180), (600, 180), 2)
    pygame.draw.line(screen, blue, (0, 240), (600, 240), 2)
    pygame.draw.line(screen, blue, (0, 300), (600, 300), 2)
    pygame.draw.line(screen, blue, (0, 360), (600, 360), 2)
    pygame.draw.line(screen, blue, (0, 420), (600, 420), 2)
    pygame.draw.line(screen, blue, (0, 480), (600, 480), 2)
    pygame.draw.line(screen, blue, (0, 540), (600, 540), 2)

    # Display whose turn it is
    if turn == 1:
        text = font.render("Player 1's Turn", True, (0, 0, 255))
        screen.blit(text, (250, 550))
    elif turn == 2:
        text = font.render("Player 2's Turn", True, (0, 0, 255))
        screen.blit(text, (250, 550))

    # Update the screen
    pygame.display.update()

# Quit Pygame
pygame.quit()

```
